```python
import pandas as pd
days_week = ("Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday")

# Keeps track of subject and the hours it was studied per day
week_log = [

]

total_hours_studied = 0.0                # Keeps track of total hours studied for the week

date = input("Enter the Date (MM/DD/YYYY): ")
print("="*10)
print()

def add_session(subject, hours):
    """Log the subject + hours studied as one variable in the week log list."""

    global total_hours_studied
    float(hours)
    total_hours_studied += hours         # Keeps track of total hours studied for the day

    entry = subject, hours               # Combines the subject + hours as one entry

    week_log.append(entry)               #Logs the entries to each day

for day in days_week:                    #To loop over however many study sessions
  subject = input(f"What subject did you study on {day}? If none, type 'None': ")
  print()
  hours = float(input(f"How many hours did you study {subject}? "))
  add_session(subject, hours)
  print("="*10)
```

```python
## Creates dataframe of the subject, hours studied, and the corresponding day
df = pd.DataFrame(week_log, columns = ['Subject', 'Hours'])
days_week_s = pd.Series(days_week, name = 'Day')
df['Day'] = days_week_s
def show_log(log):
  """Shows study sessions from each day."""
  return df

show_log(week_log)
```

```python
def subject_total(df):
  """Total hours studied per subject."""
  return df.groupby('Subject')['Hours'].sum()
print(subject_total(df))
```

```python
print("================== Summary of This Week's Study Sessions ==================")
print()
print("Study Time Per Subject: ")
print()
print(f"{show_log(week_log)}")
print()
print()
print("Hours Studied By Subject: ")
print()
print(subject_total(df))
print()
print()
print(f"Total Hours Studied: ")
print()
print(f"{total_hours_studied} hours")
print()
print("="*75)
```
